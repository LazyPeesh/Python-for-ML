```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig1, axes = plt.subplots(nrows=2, ncols=2, figsize=(4, 10), dpi=200)

axes[0, 0].plot(x, y)

axes[0, 1].plot(x, y)

axes[1, 0].plot(x, y)
axes[1, 0].set_ylabel('Y LABEL 1,0')
axes[1, 0].set_xlim(2, 6)

axes[1, 1].plot(a, b)
axes[1, 1].set_title('TITLE 1,1')

fig1.suptitle("Figure Level", fontsize=16)

plt.tight_layout()
# Manually spacing on subplots
# fig1.subplots_adjust(wspace=1, hspace=0.5)

# fig1.set_figwidth(10)

fig1.savefig('new_subplots.png', bbox_inches='tight')
```
