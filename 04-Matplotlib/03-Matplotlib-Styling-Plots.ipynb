```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')

ax.plot(x, x ** 2, label='X vs X^2')

ax.legend(loc='lower left')
ax.legend(loc=4)
ax.legend(loc=(-0.3, 0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='purple')
ax.plot(x, x, color='#e6c017', label='x vs x')  # RGB HEX Code

ax.plot(x, x + 1, color='purple', label="x vs x+1")

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#e6c017', lw=10)

ax.plot(x, x + 1, color='purple', linewidth=0.5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
# possible linestyle options '--', '-', '-.', ':', 'steps'

ax.plot(x, x, color='#e6c017', lw=10, ls='--')

ax.plot(x, x + 1, color='purple', linewidth=2, linestyle='-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

lines = ax.plot(x, x + 1, color='purple', linewidth=5)
lines[0].set_dashes([5, 2, 10, 2])  # 5 length dash / 2 space / 10 length dash / 2 space
lines[0].set_dashes(
    [1, 2, 1, 2, 10, 2])  # 1 length dash / 2 space / 1 length dash / 2 space / / 10 length dash / 2 space
```
